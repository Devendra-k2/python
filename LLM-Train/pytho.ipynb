 ## Operators in Python

Operators are special symbols or keywords used to perform operations on values and variables. Python supports several types of operators:

1. **Arithmetic Operators**: `+`, `-`, `*`, `/`, `//`, `%`, `**`
    - Used for mathematical calculations.
    - Example: `a + b`, `a * b`, `a ** b` (exponentiation)
    - result =   `5 // 2` # result is 2

2. **Comparison Operators**: `==`, `!=`, `>`, `<`, `>=`, `<=`
    - Used to compare two values.
    - Example: `a == b`, `a < b`

3. **Assignment Operators**: `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`
    - Used to assign values to variables.
    - Example: `a += 1` (equivalent to `a = a + 1`)

4. **Logical Operators**: `and`, `or`, `not`
    - Used to combine conditional statements.
    - Example: `a > 0 and b < 5`

5. **Bitwise Operators**: `&`, `|`, `^`, `~`, `<<`, `>>`
    - Used to perform bit-level operations.
    - Example: `a & b`, `a << 2`

6. **Membership Operators**: `in`, `not in`
    - Used to test if a value is present in a sequence.
    - Example: `x in list`

7. **Identity Operators**: `is`, `is not`
    - Used to compare memory locations of two objects.
    - Example: `a is b`

Operators are fundamental in writing expressions and building logic in Python programs.

## Strings in Python

A string is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). Strings are immutable, meaning their contents cannot be changed after creation.

**String Creation:**
- Single quotes: 'Hello'
- Double quotes: "World"
- Triple quotes: '''Hello World''' or """Hello World"""

**Common String Operations:**
- Concatenation: Use `+` to join strings. Example: `"Hello" + "World"` results in `"HelloWorld"`
- Repetition: Use `*` to repeat a string. Example: `"Hi" * 3` results in `"HiHiHi"`
- Indexing: Access individual characters using square brackets. Example: `s[0]` for the first character.
- Slicing: Extract substrings using `[start:end]`. Example: `s[1:4]` returns characters from index 1 to 3.

**String Methods:**
- `len(s)`: Returns the length of the string.
- `s.upper()`: Converts all characters to uppercase.
- `s.lower()`: Converts all characters to lowercase.
- `s.strip()`: Removes leading and trailing whitespace.
- `s.replace(old, new)`: Replaces occurrences of `old` with `new`.
- `s.split(separator)`: Splits the string into a list using the specified separator.
- `s.find(substring)`: Returns the index of the first occurrence of `substring`.

**Escape Characters:**
- `\n`: Newline
- `\t`: Tab
- `\\`: Backslash
- `\'`: Single quote
- `\"`: Double quote

**String Formatting:**
- f-strings: `name = "Alice"; f"Hello, {name}!"`
- `format()` method: `"Hello, {}".format(name)`

Strings are widely used for text processing, user input, and data manipulation in Python.

````python
# String examples
s = "Python Programming"
print(s[0])           # Output: P


print(s[0:6])         # Output: Python


print(s.upper())      # Output: PYTHON PROGRAMMING


print(s.lower())      # Output: python programming


print(s.replace("Python", "Java"))  # Output: Java Programming


print(len(s))         # Output: 18


print("Hello" + "World")  # Output: HelloWorld


print("Hi" * 3)       # Output: HiHiHi
````

````python
## Conditional Statements in Python

Conditional statements are used to make decisions in Python programs. They allow you to execute certain blocks of code based on whether a condition is true or false.

**Types of Conditional Statements:**
- `if` statement: Executes a block of code if the condition is true.
- `if...else` statement: Executes one block if the condition is true, another if it is false.
- `if...elif...else` statement: Checks multiple conditions in sequence.

**Syntax:**
```python
if condition:
    # code to execute if condition is true
elif another_condition:
    # code to execute if another_condition is true
else:
    # code to execute if none of the above conditions are true
```
````

````python
# Example: Conditional Statements
x = 10
if x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
    print("x is negative")

# Output: x is positive
````

## Dictionaries in Python

A dictionary is a built-in data type in Python that stores key-value pairs. Dictionaries are mutable, unordered collections where each key must be unique and immutable (e.g., string, number, tuple), but values can be of any type.

**Dictionary Creation:**
- Using curly braces: `my_dict = {"name": "Alice", "age": 25}`
- Using the `dict()` constructor: `my_dict = dict(name="Alice", age=25)`

**Accessing Values:**
- By key: `my_dict["name"]` returns `"Alice"`
- Using `get()`: `my_dict.get("age")` returns `25` (returns `None` if key not found)

**Altering Values:**
- Assign a new value: `my_dict["age"] = 26`
- Add a new key-value pair: `my_dict["city"] = "Delhi"`

**Removing Items:**
- `pop(key)`: Removes and returns value for the given key.
- `del my_dict[key]`: Deletes the key-value pair.
- `popitem()`: Removes and returns the last inserted key-value pair.
- `clear()`: Removes all items from the dictionary.

**Dictionary Methods:**
- `keys()`: Returns a view of all keys.
- `values()`: Returns a view of all values.
- `items()`: Returns a view of all key-value pairs.
- `update(other_dict)`: Updates dictionary with key-value pairs from another dictionary.

**Iteration:**
- For keys: `for key in my_dict:`
- For key-value pairs: `for key, value in my_dict.items():`

**Other Features:**
- Dictionaries support membership testing (`in`), length (`len()`), and can be nested.

Dictionaries are widely used for fast lookups, configuration, and data storage in Python programs.

````python
# Dictionary examples
my_dict = {"name": "Alice", "age": 25, "city": "Delhi"}

# Access value
print(my_dict["name"])        # Output: Alice
print(my_dict.get("age"))     # Output: 25

# Alter value
my_dict["age"] = 26
print(my_dict["age"])         # Output: 26

# Add new key-value pair
my_dict["country"] = "India"
print(my_dict)

# Remove item
my_dict.pop("city")
print(my_dict)

# Dictionary methods
print(my_dict.keys())          # Output: dict_keys(['name', 'age', 'country'])
print(my_dict.values())        # Output: dict_values(['Alice', 26, 'India'])
print(my_dict.items())         # Output: dict_items([('name', 'Alice'), ('age', 26), ('country', 'India')])

# Iteration
for key in my_dict:
    print(key, my_dict[key])

for key, value in my_dict.items():
    print(f"{key}: {value}")

# Clear dictionary
my_dict.clear()
print(my_dict)                 # Output: {}
````

## Sets in Python

A set is a built-in data type in Python that stores unordered, unique elements. Sets are mutable, but their elements must be immutable (e.g., numbers, strings, tuples). Sets are useful for removing duplicates and performing mathematical set operations.

**Set Creation:**
- Using curly braces: `my_set = {1, 2, 3}`
- Using the `set()` constructor: `my_set = set([1, 2, 3, 2])` (duplicates are removed)

**Accessing Elements:**
- Sets do not support indexing or slicing because they are unordered.
- You can iterate over a set using a loop.

**Adding Elements:**
- `add(el)`: Adds an element to the set.
- `update([el1, el2])`: Adds multiple elements.

**Removing Elements:**
- `remove(el)`: Removes the specified element (raises error if not found).
- `discard(el)`: Removes the specified element (does nothing if not found).
- `pop()`: Removes and returns an arbitrary element.
- `clear()`: Removes all elements.

**Set Operations:**
- `union(set2)`: Returns a new set with elements from both sets.
- `intersection(set2)`: Returns a new set with common elements.
- `difference(set2)`: Returns elements in the first set but not in the second.
- `symmetric_difference(set2)`: Returns elements in either set, but not both.
- `issubset(set2)`: Checks if the set is a subset of another.
- `issuperset(set2)`: Checks if the set is a superset of another.
- `copy()`: Returns a shallow copy of the set.

Sets are widely used for membership testing, removing duplicates, and mathematical operations.

````python
# Set examples
my_set = {1, 2, 3, 4, 2}
print(my_set)  # Output: {1, 2, 3, 4} (duplicates removed)

# Add elements
my_set.add(5)
my_set.update([6, 7])
print(my_set)  # Output: {1, 2, 3, 4, 5, 6, 7}

# Remove elements
my_set.remove(3)
my_set.discard(10)  # No error if 10 not present
removed = my_set.pop()  # Removes an arbitrary element
print(my_set)

# Set operations
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(set_a.union(set_b))              # Output: {1, 2, 3, 4, 5}
print(set_a.intersection(set_b))       # Output: {3}
print(set_a.difference(set_b))         # Output: {1, 2}
print(set_a.symmetric_difference(set_b)) # Output: {1, 2, 4, 5}

# Subset and superset
print(set_a.issubset({1, 2, 3, 4}))    # Output: True
print(set_a.issuperset({1, 2}))        # Output: True

# Copy and clear
copy_set = set_a.copy()
set_a.clear()
print(set_a)  # Output: set()
print(copy_set)  # Output: {1, 2, 3}
````
